```python
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import pandas_datareader.data as web
start = dt.datetime(2020,1,1)
end = dt.datetime(2024,12,5)
style.use('ggplot')
```

```python
# Microsoft (MSFT)
```

```python
df = web.DataReader('MSFT', 'stooq', start, end)
```

```python
df.to_csv('MSFT.csv')
```

```python
pd.set_option("display.precision", 2)
```

```python
df = pd.read_csv('MSFT.csv', parse_dates=True, index_col=0)
df
```

```python
df['Close'].plot()
plt.title('Adjusted Closing Price ($)');
plt.show()
```

```python
df['100ma'] = df['Close'].rolling(window=100).mean()
df
```

```python
df['100ma'].plot()
plt.title('100 Day Simple moving average');
```

```python
plt.figure(figsize = (12,6))
ax1 = plt.subplot2grid((8,1),(0,0),rowspan=5,colspan=1)
ax2 = plt.subplot2grid((8,1),(6,0),rowspan=3,colspan=1, sharex=ax1)

ax1.plot(df.index,df['Close'], 'r',label='Adjusted Close')
ax1.plot(df.index,df['100ma'], 'y', label= "100 Moving Average")
plt.xlabel('Date')

ax1.set_ylabel('Adjusted closing price ($)')
ax1.set_title('Adjusted Close and 100 day Moving Average for MSFT')
ax1.legend()

ax2.bar(df.index,df['Volume'], label = "Volume", color='b')
ax2.legend()

plt.show()
```

```python
# Oracle (ORCL)
```

```python
df = web.DataReader('ORCL', 'stooq', start, end)
```

```python
df.to_csv('orcl.csv')
```

```python
pd.set_option('display.precision', 2)
```

```python
df = pd.read_csv('orcl.csv', parse_dates=True, index_col=0)
df
```

```python
df['Close'].plot()
plt.title('Adjusted Closing Price ($)');
plt.show()
```

```python
df['100ma']=df['Close'].rolling(window=100,min_periods=0).mean()
df
```

```python
df['100ma'].plot()
plt.title('100 Day Moving Average');
```

```python
plt.figure(figsize=(12, 6))


ax1 = plt.subplot2grid((8, 1), (0, 0), rowspan=5, colspan=1)
ax2 = plt.subplot2grid((8, 1), (6, 0), rowspan=3, colspan=1, sharex=ax1)

ax1.plot(df.index, df['Close'], 'r', label='Adjusted Closing Price ($)')
ax1.plot(df.index, df['100ma'], 'g', label="100-Day Moving Average")
ax1.set_ylabel('Adjusted Closing Price ($)')
ax1.set_title('Adjusted Closing Price with 100-Day Moving Average for ORCL')
ax1.legend()


ax2.bar(df.index, df['Volume'], label="Volume", color='b')
ax2.set_ylabel('Volume')
ax2.legend()


plt.xlabel('Date')
plt.tight_layout()
plt.show()
```
